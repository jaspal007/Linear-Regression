```python
import pandas as pd
df = pd.read_csv('Tesla Dataset.csv')
df
```

```python
x = df[['Open', 'High', 'Low', 'Close', 'Adj Close']]
x
```

```python
y = df['Volume']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
pred = model.predict([[100, 100, 100, 100, 100]])
pred[0]
```

```python
import pickle
pickle.dump(model, open('model.pkl', 'wb'))
```
